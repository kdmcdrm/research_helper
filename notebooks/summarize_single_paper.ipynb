## Grobid Document Loader
This is a document loader that's supposed to be tailored to Scientific Papers. Apparently Adobe's online service will do it even more nicely, but would need an API key.

```python
from langchain.document_loaders.parsers import GrobidParser
```

```python
from langchain.document_loaders.generic import GenericLoader 
```

```python
loader = GenericLoader.from_filesystem("../papers/", 
                                       glob="*", 
                                       suffixes=[".pdf"],
                                       parser=GrobidParser(segment_sentences=False)
                                      )
```

Requires the Grobid service to be running separately.
